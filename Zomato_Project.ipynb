```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/Zomato data .csv")
print(data.head ())
```

```python
data.head()
```

```python
def handleRt(value):
     value=str(value).split('/')
     value=value[0];
     return float(value)
data['rate']=data['rate'].apply(handleRt)
print(data.head())
```

```python
data.info()
```

```python
sns.countplot(x=data['listed_in(type)'])
plt.xlabel("Type Of Restudent")
```

conclusion - majority of the resturent falls in dining category


```python
couple_data=data['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

**Conslution**- the majority of couple preferr resturent with an approximate cost of 300 rupees


```python
data.head()
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order',y = 'rate',data=data )
```

## **Conclution**-offline order received lower rateing in comparison to offline order

```python
data.head()
```

```python
import matplotlib.pyplot as plt
pivot_table = data.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table, annot = True,cmap="YlGnBu",fmt = 'd')
plt.title("heatmap")
plt.xlabel("Online Order")
plt.ylabel("list In (Type)")
plt.show()
```

###**CONCLUSION**: Dining restaurants primarily acceot office orders,whereas cafe primarily recive online orders.This suggests that clients prefence order in person at restaurants,but prefer online ordering at cafes.
